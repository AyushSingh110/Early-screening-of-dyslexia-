```python
import os
import pdfplumber
import re

# =========================
# CORRECT PATHS
# =========================
PDF_DIR = "../data/language/raw_pdfs"
OUTPUT_DIR = "../data/language/raw/normal"

print("PDF DIR exists:", os.path.exists(PDF_DIR))
print("OUTPUT DIR exists:", os.path.exists(OUTPUT_DIR))

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
print(os.listdir(PDF_DIR))
```

```python
for pdf_file in os.listdir(PDF_DIR):
    if not pdf_file.lower().endswith(".pdf"):
        continue

    pdf_path = os.path.join(PDF_DIR, pdf_file)
    extracted_text = ""

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                extracted_text += " " + page_text

    cleaned_text = re.sub(r"[^a-z\s]", "", extracted_text.lower())
    cleaned_text = re.sub(r"\s+", " ", cleaned_text).strip()

    if len(cleaned_text.split()) < 50:
        print(f"Skipped (too short): {pdf_file}")
        continue

    out_path = os.path.join(
        OUTPUT_DIR, pdf_file.replace(".pdf", ".txt")
    )

    with open(out_path, "w", encoding="utf-8") as f:
        f.write(cleaned_text)

    print(f"Saved: {out_path}")
```
